# 03 — Scale and quantities

Convert pixel lengths to **feet** using a calibration or architectural scale (dpi + feet per drawing inch).

```python
from pathlib import Path
import sys

ROOT = Path().resolve()
if not (ROOT / "src").is_dir():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT / "src"))

from blueprint_estimator.synthetic import synthetic_ingest, total_gt_length_px
from blueprint_estimator.scale_qty import ScaleConfig, total_linear_feet_segments, wall_area_sheetboard_ft2

syn = synthetic_ingest()
feet_per_px = 4.0 / 150.0
scale = ScaleConfig(feet_per_pixel=feet_per_px)
lf = total_linear_feet_segments(syn.segments_draw, scale)
print("Assumed feet_per_pixel:", feet_per_px)
print("Total linear feet (GT segments):", round(lf, 2))
print("Wall area (8 ft ceiling, rough):", round(wall_area_sheetboard_ft2(lf, 8.0), 2), "sq ft")

dpi = 150.0
scale2 = ScaleConfig(dpi=dpi, drawing_feet_per_drawing_inch=4.0)
print("Same via 1/4\"=1' style scale, feet/pixel:", scale2.resolved_feet_per_pixel())
```
